```python
# ========================================
# Exploración de Datos (EDA)
# ========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# =========================
# 2. Cargar datos
# =========================

ruta = "../01_generacion_datos/clientes_sinteticos.csv"
df = pd.read_csv(ruta)

# =========================
# 3. Exploración general
# =========================

print("Primeras filas del dataset:")
display(df.head())

print("\nResumen del dataset:")
print(df.info())

print("\nEstadísticas descriptivas:")
display(df.describe(include='all'))

# =========================
# 4. Valores nulos
# =========================

print("\nValores nulos por columna:")
print(df.isnull().sum())

# =========================
# 5. Distribución de variables
# =========================

# Variables numéricas
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
df[num_cols].hist(bins=20, figsize=(14, 10), color="skyblue", edgecolor="black")
plt.suptitle("Distribuciones de variables numéricas")
plt.tight_layout()
plt.show()

# Variables categóricas
cat_cols = df.select_dtypes(include=["object"]).columns

for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, palette="Set2")
    plt.title(f"Distribución de: {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# =========================
# 6. Correlaciones
# =========================

correlaciones = df[num_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación")
plt.show()

# =========================
# 7. Análisis específico de la variable objetivo
# =========================

# Clasificación: 'contrato_servicio' vs otras variables
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='contrató_servicio', palette='cool')
plt.title("Clientes que contrataron el servicio")
plt.show()

# Distribución por segmento de cliente
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='rubro_cliente', hue='contrató_servicio', palette='Set1')
plt.title("Segmento vs Contratación del servicio")
plt.show()

# Ingreso mensual vs contratación
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='contrató_servicio', y='presupuesto_mensual', palette='pastel')
plt.title("Ingreso mensual según contratación")
plt.show()
```
